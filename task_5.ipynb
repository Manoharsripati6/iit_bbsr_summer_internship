```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
#readint the file
df=pd.read_csv("digital_burnout.csv")
df.head() #first 5 rows
```

```python
df.columns,len(df.columns) #comlumn details
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
((df.isna().sum()/df.shape[0])*100)[df.isna().sum()>0]#percentage
```

```python
# as the null value percentage is ~ 2% , decided to impute the columns
df["social_media_hours"]=df["social_media_hours"].fillna(df["social_media_hours"].median)
df["deep_work_hours"]=df["deep_work_hours"].fillna(df["deep_work_hours"].median)
df["sleep_hours"]=df["sleep_hours"].fillna(df["sleep_hours"].median)
df["motivation_level"]=df["motivation_level"].fillna(df["motivation_level"].median)
```

```python
(df.isna().sum())[df.isna().sum()>0]
```

```python
df.info()#information about dataset
```

```python
df_cat=df.select_dtypes(include=["object","category"]).columns.tolist()
```

```python
for col in df_cat:
  print(col)
```

```python
#using label encoder for each cat col and storing the encoder object in a dict
d={}
for col in df_cat:
  le=LabelEncoder()
  df[col]=le.fit_transform(df[col].astype(str))
  d[col]=le
df, d
```
